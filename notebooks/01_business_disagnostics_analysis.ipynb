# Business Performance Diagnostic

### Project Overview

#### This project analyzes a simulated retail/e-commerce business using SQL, pandas, and ata visualization. The goal is to investigate business performance across customers, orders, products, stores, inventory, and marketing campaigns to identify potential drivers of underperformance.

### Business Context

#### The company operated across multiple regions and sales channels, with both physical store and online/mobile app orders. Leadership is concerned about overall business health and wants to better understand revenue trends, profitability, customer behavior, inventory issues, and marketing effectiveness.

### Objective

#### The objective of this analysis is to use structured data exploration to answer keyt business questions, identify performance issues, and produce clear recommendations supported by data.
